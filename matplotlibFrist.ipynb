```python
from matplotlib import pyplot as plt
import numpy as np
```

## 第一个matplotlib绘图程序

```python
x = np.arange(-50,51)
x
```

```python
#根据上述的x的值，得到该值对应的平方值
y = x ** 2
```

```python
# 绘制线性图标
plt.plot(x,y)
```

## matplotlib的方法

- 图表名称plt.title()

```python
plt.title('x ^ 2') # 默认不支持中文
plt.plot(x,y)
```

```python
# 修改字体配置以支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']
# 当设置成中文字体后，数轴上的负号将不会显示，再设置数轴上的编码方式
plt.rcParams['axes.unicode_minus'] = False;
plt.title('y是x的平方') 
plt.plot(x,y)
```

- x轴和y轴的名称

```python
# 创建x为-10 到 10的整数
x = np.arange(-10,11)
y = x ** 2
# 设置字体大小
plt.title('y = x ^ 2 x的取值范围为：[-10,10]',fontsize = 16)
plt.xlabel('X轴',fontsize = 12)
plt.ylabel('Y轴',fontsize = 12)
# 设置线条宽度
plt.plot(x,y,linewidth = 5)
```

- 绘制多个线条

```python
# 创建x为-10 到 10的整数
x = np.arange(-10,11)
y1 = x ** 2
y2 = x
# 设置字体大小
plt.title('y = x ^ 2 x的取值范围为：[-10,10]',fontsize = 16)
plt.xlabel('X轴',fontsize = 12)
plt.ylabel('Y轴',fontsize = 12)
# 设置线条宽度
plt.plot(x,y1)
plt.plot(x,y2)
```

- 设置x轴和y轴的刻度

matplotlib.pyplot.xticks(ticks=None,labels=None,**kwargs)

ticks:此参数是xtick位置的列表。是一个可选参数。如果将一个空列表作为参数，将会删除所有xticks

label:此参数包含位置再给定刻度线位置的标签。它是一个可选参数

**kwargs:此参数是文本属性，用于空置标签的外观

```python
times = ['2015/6/26','2015/8/1','2015/9/6','2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/13','2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/8/31']
# 如果是数值类型，则绘图会自动分段，如果是字符串，则会全部显示
# 随机销量
sales = np.random.randint(500,2000,size = len(times))

plt.plot(times,sales)
```

```python
# 上述图太过于密集，如果只显示不分时间，或者按照某个规则展示如何处理，就需要定制刻度
plt.xticks(range(1,len(times),2)) # 间隔一个显示一个
plt.plot(times,sales)
```

```python
plt.xticks(range(1,len(times),2),rotation = 45) # 间隔一个显示一个,rotation 将数轴的刻度倾斜
plt.plot(times,sales)
```

```python
plt.xticks(range(1,len(times),2),labels=[1,2,3,4,5],rotation = 45) # 将数轴上的刻度用其他的数值或文字代替
# lable 可以使用列表生成式来作为参数
plt.plot(times,sales)
```

- 设置图例

```python
times = ['2015/6/26','2015/8/1','2015/9/6','2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/13','2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/8/31']

# 随机出收入
income = np.random.randint(500,2000,size = len(times))
# 支出
expenses = np.random.randint(300,1500,size = len(times))

# 绘制图形
plt.xticks(range(1,len(times),2),rotation = 45)

# 注意，在使用图例前为每个图形设置label参数
plt.plot(times,income,label = "收入")
plt.plot(times,expenses,label = "支出")
# 默认会使用每个图形的label值作为图例中的说明
plt.legend() # loc参数用来设置图例说明的位置
```

- 显示每条数据的值 x,y值的位置

plt.text(x,y,string,fontsize = 15,verticalalignment="top",horizontalalignment = "right")

x,y:表示坐标值上的值
    
string：表示说明文字

fontsize：表示字体大小

verticalalignment：垂直对齐方式

horizontalalignment：水平对齐方式

```python
times = ['2015/6/26','2015/8/1','2015/9/6','2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/13','2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/8/31']

# 随机出收入
income = np.random.randint(500,2000,size = len(times))
# 支出
expenses = np.random.randint(300,1500,size = len(times))

# 绘制图形
plt.xticks(range(1,len(times),2),rotation = 45)

# 注意，在使用图例前为每个图形设置label参数
plt.plot(times,income,label = "收入")
plt.plot(times,expenses,label = "支出")
# 默认会使用每个图形的label值作为图例中的说明

plt.legend() # loc参数用来设置图例说明的位置

for x,y in zip(times,income):
    plt.text(x,y,y)
```
